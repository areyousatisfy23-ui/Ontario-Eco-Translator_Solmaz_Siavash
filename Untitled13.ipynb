```python
import os
from google.colab import userdata
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document

# 1. Activate your API Key (Ensure your secret is named GOOGLE_API_KEY)
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

# 2. Your Newmarket 2026 Data
newmarket_data = """
NT Power Residential Rates 2026 (Newmarket, Ontario):
- Ultra-Low Overnight (ULO): 3.9 cents per kWh. (11 PM to 7 AM daily).
- On-Peak: 39.1 cents per kWh. (4 PM to 9 PM weekdays).
- Mid-Peak: 15.7 cents per kWh. (7 AM to 4 PM and 9 PM to 11 PM weekdays).
- Weekend Off-Peak: 9.8 cents per kWh. (7 AM to 11 PM weekends).

IEA 2025 Energy and AI Report:
AI helps shift energy-heavy tasks to overnight periods to lower costs and stabilize the Ontario grid.
"""

# 3. Split the text
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = [Document(page_content=newmarket_data)]
split_docs = text_splitter.split_documents(docs)

# 4. Use the NEW 2026 model name: "models/gemini-embedding-001"
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

# 5. Create the Vector Database
vector_db = Chroma.from_documents(split_docs, embeddings)

print(f"вњ… Knowledge base created with {len(split_docs)} chunks of data!")
```

```python
# 1. Install the legacy support package for 2026 compatibility
!pip install -qU langchain-classic

import os
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
# We now import from langchain_classic instead of langchain
from langchain_classic.chains import RetrievalQA

# 2. Ensure your key is active
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

# 3. Setup the AI Model (Gemini Pro is a stable option)
# This code previously failed. Let's try listing models to debug.

# To debug why models are not found, you can list available models:
import google.generativeai as genai

# Make sure your API key is configured for the genai client
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

print("Available Generative AI Models:")
for m in genai.list_models():
  if "generateContent" in m.supported_generation_methods:
    print(m.name)

# If you find a suitable model, replace "gemini-pro" below with the correct model name
# For example, if you see 'gemini-pro-vision', you might use that.
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

# 4. Connect to your Newmarket database (vector_db) from Cell 2
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vector_db.as_retriever()
)

# 5. Run the query
question = "I live in Newmarket. What is the cheapest time for me to charge my car at home, and what is the exact rate?"
response = qa_chain.invoke(question)

print("-" * 30)
print(f"AI TRANSLATOR ANSWER: \n{response['result']}")
```
